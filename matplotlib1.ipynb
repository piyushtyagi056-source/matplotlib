```python
import matplotlib.pyplot as plt

months = [1,2,3,4]
sales = [1000,1500,1200,1800]

plt.plot(months , sales , color = 'green' , linestyle = '--' , linewidth = 2 , marker = 's' , label = '2026 sales data')

plt.xlabel('Months')
plt.ylabel('Sales per month')

plt.title('Monthly sales data report ')

plt.legend(loc = 'best' , fontsize = 12)

plt.grid(color = 'grey' , linestyle = ':' ,linewidth = 1)

plt.xlim(1,4)
plt.ylim(0,2000)

plt.xticks([1,2,3,4],['M1','M2','M3','M4'])

plt.show()
```

```python
#plt.bar(x , height , color = ' color name ', width = value , label = 'label name ')
import matplotlib.pyplot as plt

products = ['A', 'B' , 'C' , 'D']
sales = [1000,1500,800,1200]

plt.barh(products , sales , color = 'orange' , label = '2026 Sales' )
plt.xlabel("Products")
plt.ylabel('sales')
plt.title("product Sales comparison")
plt.legend(loc = 'best')
plt.show()
```

```python
#plt.pie(values, label= label_list , color = color_list , autopct='%1.1f%%' )
import matplotlib.pyplot as plt

region = ['North','South','East','West']
revenue =[3000,2000,1500,1000]

plt.pie(revenue,labels=region,autopct='%1.1f%%', colors=['gold','skyblue','lightgreen','coral'])
plt.title('Revenue by Region')
plt.show()
```

```python
#plt.scatter(x,y,color='color_name',marker='marker_style',label='label_name')
import matplotlib.pyplot as plt

hours_studied = [1,2,3,4,5,6,7,8]
exam_scores = [50,55,60,65,70,75,80,85]

#scatter plot
plt.scatter(hours_studied,exam_scores,color='blue' , marker='^',label='student data')

plt.xlabel('Hours Studied')
plt.ylabel('Exam Scores')
plt.title('Relationship between the Hours studied and the Exam score')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#plt.subplot(n rows, m columns, index)

import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[10,20,15,25]

plt.subplot(2,2,1)# 1 row 2 column 1 subplot
plt.plot(x,y)
plt.title('line chart')

plt.subplot(2,2,4)# 1 row 2 column 2 subplot
plt.bar(x,y)
plt.title('Bar chart')


# plt.tight_layout() ye wala mthd hamesha hi plt.show() se pehle hi aaega

plt.show()
```

```python
import matplotlib.pyplot as plt


#fig, ax=plt.subplots(n rows,m columns,figsize=(height,width))

fig, ax=plt.subplots(2,2,figsize=(10,5))

x=[1,2,3,4]
y=[10,20,15,25]

ax[0,0].plot(x,y,color='blue')
ax[0,0].set_title('Line plot')

ax[1,1].bar(x,y,color='green')
ax[1,1].set_title('Bar chart')

fig.suptitle('Comparison of  line and  Bar charts')
ax[0,1].axis('off') # to hide the plot
ax[1,0].axis('off')# to hide the plot 

plt.tight_layout()

plt.show()
```
